```python
import networkx as nx
G=nx.Graph()
v=['a','b','c','d','e']
E=[('a','b',1),('b','c',1),('a','c',1),('a','d',1),('b','e',1),('d','e',1)]
G.add_nodes_from(v)
G.add_weighted_edges_from(E)
nx.draw(G, with_labels=True)
```

```python
d=nx.greedy_color(G,strategy="random_sequential")
color_name = ['green', 'lightblue', 'red', 'ivory', 'lavender', 'grey' ]
sortedDict = dict( sorted(d.items()) )
m = []
for k,v in sortedDict.items():
    m.append(color_name[v])
l=list(d.values())
print("chromatic number is", max(l)+1)
nx.draw(G,node_size=1000,node_color=m, with_labels=True)
```
